## Informacion y visualizacion de los datos: ##

```python
dfEdx.groupby(['course_type'])['course_type'].count()
```

```python
plt.figure(figsize = (6, 6))
plt.pie(dfEdx.groupby(['course_type']).size(),
        labels = ['Instructor-led on a course schedule', 'Self-paced on your time'], 
        autopct = '%1.1f%%',
        startangle = 90, colors = ['#facd75','#fd6767'],
        textprops = {'size': 16},
        wedgeprops= {"edgecolor":"white",
                     'linewidth': 1,
                     'antialiased': True})
plt.axis('equal')
plt.title('Course types:\n',loc='left', fontdict={'fontsize': 18})
plt.show()
```

```python
print(pd.unique(dfEdx['language'].sort_values()))
```

```python
dfEdx.replace('اللغة العربية', 'Arabe', inplace=True)
dfEdx.replace('中文','Chino', inplace=True)
dfEdx.replace('日本語','Japones', inplace=True)
print(pd.unique(dfEdx['language'].sort_values()))
```

```python
plt.rcParams.update({'font.size': 12, 'axes.axisbelow': True})
plt.figure(figsize = (10, 6))
plt.hist(dfEdx['language'],bins=50,range=(0,9), rwidth=0.9,
         color='#fd6767')
plt.grid(axis='y', color ='Grey',
        linestyle ='-.', linewidth = 0.1)
plt.xticks(range(0,5,1))
plt.margins(0.01)
plt.xlabel("Idioma", labelpad=10)
plt.ylabel("Cantidad")
plt.title('Distribucion del lenguaje por cursos:\n',loc='center', fontdict={'fontsize': 16})
plt.show()
```

```python
import matplotlib.ticker as ticker


a_df = dfEdx.groupby(['course_length']).size().sort_values(ascending=True)
fig, ax = plt.subplots(figsize=(8, 6)) 
ax.barh(a_df.index, a_df, color='#1a3f5f')
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x/1)))
plt.margins(0.01)
plt.xlabel("The number of course", labelpad=10)
plt.grid(axis='x', color ='Grey',
        linestyle ='-.', linewidth = 0.1)
plt.suptitle('The number of courses for each category:',fontsize= 16)
plt.show()
```

```python
a_df = dfEdx.groupby(['course_effort']).size().sort_values(ascending=True)
fig, ax = plt.subplots(figsize=(8, 20)) 
ax.barh(a_df.index, a_df, color='#1a3f5f')
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x/1)))
plt.margins(0.01)
plt.xlabel("The number of course", labelpad=6)
plt.grid(axis='x', color ='Grey',
        linestyle ='-.', linewidth = 0.1)
plt.title('Esfuerzo semanal por curso:',fontsize= 16)
plt.show()
```

```python
a_df = dfEdx.groupby(['Level']).size().sort_values(ascending=True)
fig, ax = plt.subplots(figsize=(8, 1)) 
ax.barh(a_df.index, a_df, color='#1a3f5f')
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x/1)))
plt.margins(0.01)
plt.xlabel("The number of course", labelpad=10)
plt.grid(axis='x', color ='Grey',
        linestyle ='-.', linewidth = 0.1)
plt.title('Cantidad de cursos por nivel:',fontsize= 16)
plt.show()
```

```python
a_df = dfEdx.groupby(['subject']).size().sort_values(ascending=True)
fig, ax = plt.subplots(figsize=(8, 12)) 
ax.barh(a_df.index, a_df, color='#1a3f5f')
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '{:,.0f}'.format(x/1)))
plt.margins(0.01)
plt.xlabel("The number of course", labelpad=6)
plt.grid(axis='x', color ='Grey',
        linestyle ='-.', linewidth = 0.1)
plt.title('Cantidad de cursos por area:',fontsize= 16)
plt.show()
```

```python
dfEdx2=dfEdx
dfEdx2['n_enrolled'] = dfEdx2['n_enrolled'].astype(pd.StringDtype())
dfEdx2['n_enrolled'] = dfEdx2['n_enrolled'].fillna('0')
dfEdx2['n_enrolled'] = pd.to_numeric(dfEdx2['n_enrolled'].str.replace(',', ''))
dfEdx2['n_enrolled']
```

```python
e_df = dfEdx2.groupby('institution')['n_enrolled'].sum()
print('Top 15 Institutos por mas subscriptores:')
e_df = e_df.sort_values(ascending=False)
e_df.head(15).astype(int)
```

```python
e_df = dfEdx2.groupby('title')['n_enrolled'].sum()
print('Top 15 cursos por mas subscriptores:')
e_df = e_df.sort_values(ascending=False)
e_df.head(15).astype(int)
```

```python
e_df = dfEdx2.groupby('course_type')['n_enrolled'].sum()
print('Cantidad de subscriptores por tipo de curso:')
e_df = e_df.sort_values(ascending=False)
e_df.head(18).astype(int)
```

```python
e_df = dfEdx2.groupby('Level')['n_enrolled'].sum()
print('Cantidad de subscriptores por nivel:')
e_df = e_df.sort_values(ascending=False)
e_df.head(18).astype(int)
```

```python
e_df = dfEdx2.groupby('subject')['n_enrolled'].sum()
print('Cantidad de subscriptores por tópico:')
e_df = e_df.sort_values(ascending=False)
e_df.head(18).astype(int)
```

```python
e_df = dfEdx2.groupby('language')['n_enrolled'].sum()
print('Cantidad de subscriptores por idioma:')
e_df = e_df.sort_values(ascending=False)
e_df.head(18).astype(int)
```

```python
e_df.plot(kind='barh')
```

```python
dfEdx2['price'] = dfEdx2['price'].str.replace("FREE-Add a Verified Certificate for $","",regex=False)
print(dfEdx2['price'].head(2))
dfEdx2['price'] = dfEdx2['price'].str.replace(" USD","",regex=False)
print(dfEdx2['price'].head(2))
```

```python
dfEdx2['price'] = dfEdx2['price'].astype(pd.StringDtype())
dfEdx2['price'] = dfEdx2['price'].replace('FREE-Add a Verified Certificate for ',' ')
e_df = dfEdx2.groupby('price')['n_enrolled'].sum()
print('Cantidad de subscriptores por precio:')
e_df = e_df.sort_values(ascending=False)
e_df.head(18).astype(int)
```

```python

dfEdx3 = dfEdx2
dfEdx3['price'] = dfEdx3['price'].astype(float)
plt.rcParams.update({'font.size': 12, 'axes.axisbelow': True})
plt.figure(figsize = (10, 6))
plt.hist(dfEdx3['price'],bins=50,range=(0,1000), rwidth=0.85,
         color='#fd6767')
plt.grid(axis='y', color ='Grey',
        linestyle ='-.', linewidth = 0.1)
plt.xticks(range(0,1001,100))
plt.margins(0.01)
plt.xlabel("Precio", labelpad=10)
plt.ylabel("Cantidad")
plt.title('Distribucion del precio de los cursos:\n',loc='center', fontdict={'fontsize': 16})
plt.show()
```

En este data set se pueden apreciar varias preferencias sobre los cursos: <br>
* En relación al costo la mayor cantidad de inscriptos eligen cursos con un precio inferior a los 100usd.
